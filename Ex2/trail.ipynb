```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("house_pred.csv")
data.head()
```

```python
target = data['SalePrice']
data = data.drop(['Id','Alley','PoolQC','Fence','MiscFeature','SalePrice'],axis=1)
```

```python
target
```

```python
obj = data.select_dtypes(include=object).columns
num = data.select_dtypes(exclude=object).columns
```

```python
for i in num:
    data[i].fillna(data[i].mean(), inplace=True)
```

```python
for i in data.columns[data.isna().sum()>0]:
    data[i].fillna(None, method='ffill',inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

xtr, xtst, ytr, ytst = train_test_split(data, target, test_size=0.3)

lr = LinearRegression()
```

```python
model = lr.fit(xtr, ytr)
```

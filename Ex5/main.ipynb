```python
import pandas as pd
import numpy as np
```

```python
data= pd.read_csv('/home/kavin/Kavin/sem6/ML/Lab/Ex5/Telco-Customer-Churn.csv')
data
```

```python
target = data['Churn']
data = data.drop(['customerID','Churn'],axis=1)
```

```python
data['TotalCharges'] = data['TotalCharges'].replace(' ',np.nan).astype(float)
```

```python
obj = data.select_dtypes(include=object).columns
num = data.select_dtypes(exclude=object).columns
obj
```

```python
z = list(num)[1:]
z
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
sc.fit(data[z])
data[z] = sc.transform(data[z])
```

```python
data
```

```python
data
```

```python
for i in num:
    data[i].fillna(data[i].mean(), inplace=True)
```

```python
for i in data.columns[data.isna().sum()>0]:
    data[i].fillna(None, method='ffill',inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()

for i in data.select_dtypes(include=object):
        data[i]= enc.fit_transform(data[i])

target = enc.fit_transform(target)
```

```python
data.columns
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

xtr, xtst, ytr, ytst = train_test_split(data, target, test_size=0.3)

lr = LogisticRegression()
```

```python
lr.fit(xtr, ytr)
lr.score(xtr, ytr)
```

```python
from sklearn.neighbors import  KNeighborsClassifier
KNN = KNeighborsClassifier()
```

```python
KNN.fit(xtr,ytr)
KNN.score(xtst,ytst)
```

```python
from sklearn.naive_bayes import GaussianNB
nv = GaussianNB()
```

```python
nv.fit(xtr,ytr)
nv.score(xtst,ytst)
```

```python
from sklearn.tree import  DecisionTreeClassifier
d_tree = DecisionTreeClassifier()
```

```python
d_tree.fit(xtr,ytr)
d_tree.score(xtst,ytst)
```

```python
from sklearn.svm import SVC

svc = SVC()
```

```python
svc.fit(xtr,ytr)
svc.score(xtst,ytst)
```

| Model         | Accuracy     |
|--------------|-----------|
| Logistic Regression | 79.8%     |
| KNN      | 76.05%  |
| Naive Bayes      | 75.6%  |
| Decision Trees      | 73.9%  |
| SVM      | 81.1%  |

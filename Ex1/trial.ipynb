```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
x = np.random.rand(1000, 1)
x.shape
```

## data

```python
m = 5
c = 3
y = m*x + c + np.random.normal(2, 1, x.shape)
plt.scatter(x, y)
```

## equation

```python
m = np.sum((x - x.mean()) * (y - y.mean())) / np.sum((x-x.mean())**2)
c = y.mean() - m*x.mean()
plt.scatter(x, y)
plt.plot(x, m*x+c, c="red")
m, c
```

## sklearn

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x, y)
m = float(model.coef_)
c = float(model.intercept_)
plt.scatter(x, y)
plt.plot(x, m*x+c, c="red")
m, c
```

## random lines

```python
plt.scatter(x, y)
min_error = float('inf')

for i in range(1000):
    slope = np.random.randint(1,10) * np.random.random()
    inter = np.random.randint(1,10) *np.random.random()

    y1 = slope*x + inter
    error = ((y - y1)**2).mean()
    if error< min_error:
        min_error = error
        m = slope
        c = inter

plt.plot(x, m*x+c, c="red")
m,c
```

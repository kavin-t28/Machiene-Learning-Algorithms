```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
array = np.random.rand(1500,2)
rand_array = np.random.rand(1500,2)
array[:, 1] = 2 * array[:, 0] + 5 + np.random.randn(1500) * 0.4  # standard deviation is 0.4

array
```

```python
train_data = array[:1000, :]
test_data = array[1000:, :]
```

```python
lines_of_fit  = []
errors=[]
```

```python
for i in range(100):
    m,b=np.polyfit(train_data[:,0],train_data[:,1],1)
    lines_of_fit.append((m,b))


    error= np.mean((train_data[:,1]- (m * train_data[:, 0] + b))**2)
    errors.append(error)
```

```python
best_fit_index = np.argmin(errors)
best_fit_line = lines_of_fit[best_fit_index]
```

```python
predictions = best_fit_line[0] * test_data[:, 0] + best_fit_line[1]
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(train_data[:, 0], train_data[:, 1], s=4, label='Training Data')
plt.scatter(test_data[:, 0], test_data[:, 1], s=4, label='Testing Data')
plt.plot(test_data[:, 0], predictions, color='red', linewidth=2, label='Line of Best Fit')
plt.title("Data Distribution and Line of Best Fit")
plt.legend()
plt.show()
```

```python
print("Line of best fit: y = {:.4f}x + {:.4f}".format(*best_fit_line))
print("Predictions for the testing set:", predictions)
```
